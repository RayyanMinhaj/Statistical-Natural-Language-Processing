Consider a setting where the task is to classify a collection of scientific articles based on their topics. The table below shows some word occurrence statistics from this document collection for four topic categories: mathematics, chemistry, astronomy, and physics.

|                 | N(composition) | N(gravity) | N(differential) | N(theory) |
|-----------------|----------------|------------|-----------------|-----------|
| c = mathematics | 30             | 3          | 50              | 7         |
| c = chemistry   | 43             | 0          | 2               | 5         |
| c = astronomy   | 47             | 53         | 19              | 11        |
| c = physics     | 30             | 34         | 29              | 7         |

1. Explain how we can use $\chi^2$ statistics to asses wether each word feature would be useful for predicting the class. (0.5 Points)
2. Implement the function chi2_statistic and compute the $\chi^2$ statistics for the given features. (0.5 Points)

### Answers 
1. chi-squared basically measures degree of association between class labels and a word's occurrence. If a word's occurrence varies greatly across classes, it suggests the word is associated with the class, and is informative. If the word appears uniformly across all classes, it's likely uninformative. We use chi-squared for feature selection, High chi-squared values suggest that the word is strongly associated with at least one class → Good feature. Low chi-squared values suggest the word is independent of the class → Poor feature. You can then rank words by their chi-squared scores and select the top k as features for a classifier like Naive Bayes or SVM.<br/>
$$X^2(w) = \sum \frac{(O_{w,c}-E_{w,c})^2}{E_{w,c}}$$

```python
import numpy as np
contingency_table = np.array([[30 ,3 ,50,7],
                              [43,0,2,5],
                              [47,53,19,11],
                              [30,34,29,7]])
```

```python
row_sums = np.sum(contingency_table, axis=1)
col_sums = np.sum(contingency_table, axis=0)

print("Row sum: ", row_sums)

print("Col sum: ", col_sums)
```

```python
def chi2_statistic(contingency_table: list[list[int]]) -> float:
    """
    Compute the chi-squared statistic from a contingency table.

    Parameters:
        contingency_table (list of list of int): 
            A 2D table where entry [i][j] is the count of samples with 
            feature value i and label j.

    Returns:
        float: The chi-squared statistic.
    """

    total = np.sum(contingency_table)
    row_sums = np.sum(contingency_table, axis=1)
    col_sums = np.sum(contingency_table, axis=0)

    chi2_values = []

    for j in range(contingency_table.shape[1]):
        chi2_col = 0.0
        for i in range(contingency_table.shape[0]):
            observed = contingency_table[i, j]
            expected = (row_sums[i] * col_sums[j]) / total
            if expected > 0:
                chi2_col += (observed - expected)**2 / expected
        chi2_values.append(chi2_col)

    #now we need to take chi2_avg for the values in chi2_values (every word)
    chi2_avg = 0
    for i in range(0, len(chi2_values)):
        chi2_avg += chi2_values[i] * (row_sums[i]/sum(row_sums)) #this is prior
        
    
    return chi2_avg
```

```python
print(f'Chi-Squared Statistic for given Table: {chi2_statistic(contingency_table)}')
```

1. Define the states and observations from the training data and compute the initial probabilities, i.e the probability of starting in a state. (1 Point)

```python
lines = open("data/eng.train").readlines()

sentences = []
temp = []

for line in lines:
    
    if line == "\n":
        sentences.append(temp)
        temp = []
        continue
        
    toks = line.split()

    temp.append((toks[0], toks[3]))

print(sentences[:5])
```

```python
#extract states and observations
states = set()
observations = set()

for sent in sentences:
    for word, tag in sent:
        states.add(tag)
        observations.add(word)

#alculatimg initial probabilities
init_counts = {state: 0 for state in states}
total_sentences = len(sentences)

for sent in sentences:
    first_state = sent[0][1]  #tag of first word in sentence
    init_counts[first_state] += 1


init_probs = {state: count / total_sentences for state, count in init_counts.items()}

print("States:", states, "\n")

print("Number of sentences:", total_sentences, "\n")

print("Initial probabilities (start of sentence):")
for state, prob in init_probs.items():
    print(f"{state}: {prob:.4f}")
```

2. Compute the probability of transitioning from one state to another $ P(\pi_i|\pi_{i-1})$. Define `<start>` as the start state and `<end>` as the end state i.e. start and end of each sentence when computing the transition counts. (1 Point)

```python
from collections import defaultdict

transition_counts = defaultdict(lambda: defaultdict(int))
for sent in sentences:
    tags = [tag for _, tag in sent]
    transition_counts["<start>"][tags[0]] += 1 #first tag

    for i in range(1, len(tags)):
        prev_tag = tags[i - 1]
        curr_tag = tags[i]
        transition_counts[prev_tag][curr_tag] += 1

    
    transition_counts[tags[-1]]["<end>"] += 1 #add last tag

#convert counts to probabilities
transition_probs = defaultdict(dict)

for prev_tag in transition_counts:
    total = sum(transition_counts[prev_tag].values())
    for curr_tag in transition_counts[prev_tag]:
        transition_probs[prev_tag][curr_tag] = transition_counts[prev_tag][curr_tag] / total

#Sample output 
print("Sample Transition Probabilities:")
for prev_tag in list(transition_probs)[:3]:
    for curr_tag in transition_probs[prev_tag]:
        print(f"P({curr_tag} | {prev_tag}) = {transition_probs[prev_tag][curr_tag]:.4f}")
```

3. Compute the emission probabilities, i.e. the probability of seeing a observation in a given state. (1 Point)

```python
from collections import defaultdict

emission_counts = defaultdict(lambda: defaultdict(int))
tag_counts = defaultdict(int)

for sent in sentences:
    for word, tag in sent:
        emission_counts[tag][word] += 1
        tag_counts[tag] += 1  #total number of times tag appears

emission_probs = defaultdict(dict)

for tag in emission_counts:
    total = tag_counts[tag]  
    for word in emission_counts[tag]:
        emission_probs[tag][word] = emission_counts[tag][word] / total

#printing a few emission probabilities for inspection
print("Sample Emission Probabilities:")
for tag in list(emission_probs.keys())[:3]:
    for word in list(emission_probs[tag].keys())[:3]:
        print(f"P({word} | {tag}) = {emission_probs[tag][word]:.4f}")
```

# Exercise 5 Conditional random fields 3 Points
1. While HMMs are generative models CRFs are discriminativ. Explain why? (0.5 points)
2. What are the advantages of CRFs over HMMs? (0.5 points)
   
Now you will use the `sklearn_crfsuite` to implement a CRF for Part of Speech tagging. You'll be using the Treebank corpus for that. Run the following cell to get the data.

### Answers
1. HMMs are generative models because they model the joint probability of observations and labels -> P(x,y) = P(y) * P(x|y). They assume a generative process, basically how the data (observations) is generated from the hidden states. CRFs (Conditional Random Fields) are discriminative models because they directly model the conditional probability of labels given observations -> P(y|x) They focus only on the boundaries between classes without assuming how the observations are generated. In short, HMMs model how data is generated (P(x,y)) and CRFs model how to classify data (P(y|x))

2. CRFs allow the use of entire observation sequences and multiple features (like word shape, neighboring words, capitalization) without making independence assumptions.\
CRFs normalize over the entire sequence, avoiding label bias problem (where states with fewer outgoing transitions tend to get unfairly favored).\
CRFs can incorporate arbitrary and overlapping features of the input data (like word suffixes, previous words, part-of-speech tags, etc.)\
CRFs usually achieve higher accuracy in tasks like Named Entity Recognition, POS tagging, etc., due to their flexibility and feature support.

```python
import sklearn_crfsuite
from sklearn_crfsuite import metrics

train_sents= corpus

def word2features(sent, i):
    """
    Generate features for a given word in a sentence.

    Args:
    sent (list): A list of tuples representing a sentence, where each tuple contains a word and its label.
    i (int): The index of the word in the sentence for which features are to be generated.

    Returns:
    dict: A dictionary where keys are feature names and values are corresponding feature values.
    """
    word = sent[i][0]

    features = {
        'word': word,
        'is_first': i == 0,
        'is_last': i == len(sent) - 1,
        'is_capitalized': word[0].isupper(),
        'is_all_caps': word.isupper(),
        'is_all_lower': word.islower(),
        'prefix-1': word[0],
        'prefix-2': word[:2],
        'prefix-3': word[:3],
        'suffix-1': word[-1],
        'suffix-2': word[-2:],
        'suffix-3': word[-3:],
        'prev_word': '' if i == 0 else sent[i - 1][0],
        'next_word': '' if i == len(sent) - 1 else sent[i + 1][0],
        'has_hyphen': '-' in word,
        'is_numeric': word.isdigit(),
        'has_capitals_inside': any(c.isupper() for c in word[1:]),
    }

    return features


def sent2features(sent):
    """
    Generate features for all words in a sentence.

    Args:
    sent (list): A list of tuples representing a sentence, where each tuple contains a word and its label.

    Returns:
    list: A list of feature dictionaries. Each dictionary corresponds to the features of a word in the sentence.
    """
    features = []

    for i in range(len(sent)):
        features.append(word2features(sent, i))
    
    return features


def sent2labels(sent):
    """
    Extract labels from a sentence.

    Args:
    sent (list): A list of tuples representing a sentence, where each tuple contains a word and its label.

    Returns:
    list: A list of labels extracted from the sentence.
    """
    labels = []
    for _, label in sent:
        labels.append(label)
    
    return labels
```

4. Train the CRF and run the evaluation code. For that look up the documentation of [sklearn_crfsuite](https://sklearn-crfsuite.readthedocs.io/en/latest/tutorial.html#training) on how to set up the training code. (1 point).

```python
#Train the CRF
X_train = [] 
y_train = []
X_test = []
y_test = []

for s in corpus:
    X_train.append(sent2features(s))

for s in corpus:
    y_train.append(sent2labels(s))

for s in test_data:
    X_test.append(sent2features(s))

for s in test_data:
    y_test.append(sent2labels(s))


crf = sklearn_crfsuite.CRF(
    algorithm='lbfgs',
    c1=0.1,           
    c2=0.1,           
    max_iterations=100,
    all_possible_transitions=True
)


crf.fit(X_train, y_train)
```

```python
#Evaluation code
y_pred = crf.predict(X_test)
labels=list(crf.classes_)

f1 = metrics.flat_f1_score(y_test, y_pred, average='weighted', labels=labels)

print("Weighted f1 score: ",f1)
```
